```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/eddieloyd/DIS/BigData/final/Big-Data-March-Madness/data/all years.csv")
df = df.rename(columns = {"Tourament-Result": "Tournament-Result"})
```

```python
# 1992, 2002, 2012, 2018
# df
```

```python
# tournament teams
tournament_teams = df.loc[df["Tournament-Result"] > 0]
# tournament_teams
```

```python
non_tournament = df.loc[df["Tournament-Result"] == 0]
# non_tournament
```

```python
# W, L, W-L%, SRS, SOS, FG, FGA, Pts-Tm., Pts-Opp., MP,FG,FGA,FG%,3P,3PA,3P%,FT,FTA,FT%,ORB,TRB,AST,STL,BLK,TOV,PF,Pace,ORtg,FTr,3PAr,TS%,TRB%,AST%,STL%,BLK%,eFG%,TOV%,ORB%,FT/FGA
```

```python
# 1992 plot
tourny_teams_92 = tournament_teams.loc[tournament_teams["Season"].str.contains("1992-93")]
non_tourny_92 = non_tournament.loc[non_tournament["Season"].str.contains("1992-93")]

#Need to figure out the legend still

# handles, labels = plt.gca().get_legend_handles_labels()
# fig.legend(["Green", "Blue"], ["Non-Tournament", "Tournament"], loc='upper center')

# plt.legend(["Non-tournament", "Tournament"], loc = "upper right")


# axs[0,0].ast_tourny_92.plot.density(color = "green")
# ast_non_92.plot.density(color = "blue")
# plt.title("Assist%")
```

```python
fig, axs = plt.subplots(7, 5, figsize = (24, 24))
stats = ["W", "L", "W-L%", "SRS", "SOS", "FG", "FGA", 
             "Pts-Tm.", "Pts-Opp.", "MP","FG","FGA","FG%","3P","3PA","3P%",
             "FT","FTA","FT%","ORB","TRB","AST","STL","BLK","TOV","PF",
             "FTr","3PAr","TS%","TRB%","AST%","BLK%","eFG%","TOV%","FT/FGA"]
plot_x = 0
plot_y = 0
for stat in stats:

    ast_tourny_92 = tourny_teams_92[stat]
    ast_non_92 = non_tourny_92[stat]
    sns.kdeplot(ast_tourny_92, color = "green", ax = axs[plot_x,plot_y])
    sns.kdeplot(ast_non_92, color = "blue", ax = axs[plot_x,plot_y])
    #axs[plot_x, plot_y].set_title(stat)
    
    plot_y += 1
    if plot_y %5 ==0:
        plot_x += 1
        plot_y = 0
    
# Green is Tournament, Blue is non Tournament
# Still need to figure out legend
```
